### Sine Wave Formula
$c\sin(ax+d)+b$

```python
def plot_3d_np(matrix):
    dim = len(matrix)
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    color_lst = []
    point_lst = [i for i in range(len(matrix))]
    for x in range(dim):
        for y in range(dim):
            for z in range(dim):
                color = matrix[x, y, z]
                if color == 0:
                    color_lst.append("blue")
                elif color == 1:
                    color_lst.append("brown")
                else:
                    color_lst.append("white")
    ax.scatter([t[0] for t in product(point_lst, point_lst, point_lst)], [t[2] for t in product(point_lst, point_lst, point_lst)], [t[1] for t in product(point_lst, point_lst, point_lst)], c = color_lst)
    plt.show
def gen_mc_cube(dim):
    rand_int = random.randint(1, 10)
    rand_int1 = random.randint(1, 10)
    rand_int2 = random.randint(1, 10)
    z = np.zeros(shape=(dim,dim,dim))
    woot = np.random.randint(0, 203492523542463)
    colors = ['red', 'orange', 'yellow', 'green','blue', 'magenta']
    for x in range(dim):
        z[:x, :x, :x] += 1
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    point_lst = [i for i in range(dim)]
    pt = product(point_lst, point_lst, point_lst)
    lst_color = []
    x_lst = []
    y_lst = []
    z_lst = []
    for x in z:
        for y in x:
            for zval in y:
                    lst_color.append(colors[int((woot * np.sin(zval))%5)])
    for tup in product(point_lst, point_lst, point_lst):
        x_lst.append(tup[0])
        y_lst.append(tup[1])
        z_lst.append(tup[2])
    ax.scatter(x_lst, y_lst, z_lst, color = lst_color, s = 200, c = lst_color)
    plt.show()
def gen_3c_cube(dim, color1, color2, color3):
    z = np.zeros(shape=(dim,dim,dim))
    for x in range(dim):
        z[:x, :x, :x] += 1
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    point_lst = [i for i in range(dim)]
    pt = product(point_lst, point_lst, point_lst)
    lst_color = []
    x_lst = []
    y_lst = []
    z_lst = []
    for x in z:
        for y in x:
            for zval in y:
                if zval%3 == 0:
                    lst_color.append(color1)
                elif zval%3 == 1:
                    lst_color.append(color2)
                else:
                    lst_color.append(color3)
    for tup in product(point_lst, point_lst, point_lst):
        x_lst.append(tup[0])
        y_lst.append(tup[1])
        z_lst.append(tup[2])
    ax.scatter(x_lst, y_lst, z_lst, color = lst_color, s = 200, c = lst_color)
    plt.show()
def gen_2c_cube(dim, color1, color2):
    z = np.zeros(shape=(dim,dim,dim))
    for x in range(dim):
        z[:x, :x, :x] += 1
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    point_lst = [i for i in range(dim)]
    pt = product(point_lst, point_lst, point_lst)
    lst_color = []
    x_lst = []
    y_lst = []
    z_lst = []
    for x in z:
        for y in x:
            for zval in y:
                if zval%2 == 0:
                    lst_color.append(color1)
                else:
                    lst_color.append(color2)
    for tup in product(point_lst, point_lst, point_lst):
        x_lst.append(tup[0])
        y_lst.append(tup[1])
        z_lst.append(tup[2])
    ax.scatter(x_lst, y_lst, z_lst, color = lst_color, s = 200, c = lst_color)
    plt.show()
def generate_rand_sin(size):
    return np.sin(np.random.multinomial(360, [1/size]*size, size=1))
def gen_christmas_cube(dim):
    rand_int = random.randint(1, 10)
    rand_int1 = random.randint(1, 10)
    rand_int2 = random.randint(1, 10)
    z = np.zeros(shape=(dim,dim,dim))
    woot = np.random.randint(0, 203492523542463)
    colors = ['red', 'green', 'white']
    for x in range(dim):
        z[:x, :x, :x] += 1
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    point_lst = [i for i in range(dim)]
    pt = product(point_lst, point_lst, point_lst)
    lst_color = []
    x_lst = []
    y_lst = []
    z_lst = []
    for x in z:
        for y in x:
            for zval in y:
                    lst_color.append(colors[int((woot * np.sin(zval))%3)])
    for tup in product(point_lst, point_lst, point_lst):
        x_lst.append(tup[0])
        y_lst.append(tup[1])
        z_lst.append(tup[2])
    ax.scatter(x_lst, y_lst, z_lst, color = lst_color, s = 200, c = lst_color)
    plt.show()
```
